```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


mean = 5
std_dev = 2


x_normal = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
y_normal = stats.norm.pdf(x_normal, mean, std_dev)


mean_std_normal = 0
std_dev_std_normal = 1
x_std_normal = np.linspace(-4, 4, 100)
y_std_normal = stats.norm.pdf(x_std_normal, mean_std_normal, std_dev_std_normal)


plt.figure(figsize=(12, 6))


plt.plot(x_normal, y_normal, label=f'Normal Distribution (mean={mean}, std_dev={std_dev})')


plt.plot(x_std_normal, y_std_normal, label='Standard Normal Distribution (mean=0, std_dev=1)', linestyle='dashed')

plt.title('Normal Distribution vs. Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def calculate_z_score(x, mean, std_dev):
    z = (x - mean) / std_dev
    return z


score = 85
mean = 70
std_dev = 10

z_score = calculate_z_score(score, mean, std_dev)
print(f"The Z-score is: {z_score}")
```
